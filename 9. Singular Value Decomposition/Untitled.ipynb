```python
import cv2
import numpy as np
from numpy.linalg import svd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
A = np.array([[1,0,0,0],[0,0,0,4],[0,3,0,0],[0,0,0,0],[2,0,0,0]])
A
```

```python
U,s,VT = svd(A)
```

```python
U,s,VT
```

```python
cimg = cv2.imread(r'.\images\image01.jpg')
print(cimg.shape)
plt.imshow(cv2.cvtColor(cimg,cv2.COLOR_BGR2RGB))
```

```python
img = cv2.cvtColor(cimg,cv2.COLOR_BGR2GRAY)
print(img.shape)
plt.imshow(img,cmap='gray')
```

```python
U,s,VT = svd(img)
U.shape,s.shape,VT.shape
```

```python
num = 100
re_img = U[:,:num].dot(np.diag(s[:num])).dot(VT[:num,:])
re_img.shape
plt.imshow(re_img,cmap='gray')
```
