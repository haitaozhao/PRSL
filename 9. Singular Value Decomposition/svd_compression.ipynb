# 运用SVD进行图像压缩

```python
# 导入相关包
import cv2
import numpy as np
from numpy.linalg import svd
import matplotlib.pyplot as plt
%matplotlib inline
```

## 读取图片

```python
# 读取图片
cimg = cv2.imread(r'.\images\image01.jpg')
print(cimg.shape)
plt.imshow(cv2.cvtColor(cimg,cv2.COLOR_BGR2RGB))
```

```python
# 灰度图片
img = cv2.cvtColor(cimg,cv2.COLOR_BGR2GRAY)
print(img.shape)
plt.imshow(img,cmap='gray')
```

## 奇异值分解

```python
U,s,VT = svd(img)
U.shape,s.shape,VT.shape
```

```python
plt.plot(np.log(s))
```

## 图像重建

```python
num = 400
re_img = U[:,:num].dot(np.diag(s[:num])).dot(VT[:num,:])
re_img.shape
```

```python
plt.imshow(re_img,cmap='gray')
```

```python
## 重建误差
error = np.sqrt(np.sum((img-re_img)**2)) # F-norm
error2 = np.sqrt(sum(s[num:]**2))
```
